## 3. Analysis Questions

__This section answers key questions related to student performance.__


### Average Final Grade (G3)


```python
import pandas as pd

df = pd.read_csv('student-mat.csv', sep=";")

#average grade (G3)
average_g3 = df["G3"].mean()

print("Average final grade (G3):",average_g3)
```

### Number of Students Scoring Above 15


```python
#count of students who scored above 15 marks
count = (df['G3']>15).sum()
print(count)
```

### Relationship Between Study Time and Performance


```python
# correlation between study time and final grade
import pandas as pd

df = pd.read_csv("student-mat.csv", sep=";")

# Select relevant columns
data = df[['studytime', 'G1', 'G2', 'G3']]

# Show first few rows
print(data.head())

# Calculate correlation
correlation = data.corr()

print("\nCorrelation Matrix:")
print(correlation)
```

__The correlation value indicates a weak positive relationship between study time and final grades.__

### Gender-wise Performance Comparison

```python
# average final grade by gender
avg_gender = df.groupby('sex')["G3"].mean()

if avg_gender['M'] > avg_gender['F']:
    print("Male students perform better on average than female students.\n")
else:
    print("Female students\n")
    
print(avg_gender)
```

## 4. Visualizations

The following plots help visualize grade distribution and performance patterns.


```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("student-mat.csv", sep=";")

# Plot histogram of final grades
plt.hist(df['G3'], bins=10)
plt.xlabel("Final Grade (G3)")
plt.ylabel("Number of Students")
plt.title("Histogram of Final Grades")
plt.show()
```

### Scatter Plot: Study Time vs Grades


```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("student-mat.csv", sep=";")

# Scatter plot
plt.scatter(df['studytime'], df['G3'])
plt.xlabel("Study Time")
plt.ylabel("Final Grade (G3)")
plt.title("Study Time vs Final Grade")
plt.show()
```

### Bar Chart: Male vs Female Average Score


```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("student-mat.csv", sep=";")

# Calculate average G3 score by gender
avg_scores = df.groupby('sex')['G3'].mean()

# Bar chart
avg_scores.plot(kind='bar')
plt.xlabel("Gender")
plt.ylabel("Average Final Grade (G3)")
plt.title("Male vs Female Average Final Score")
plt.show()
```
